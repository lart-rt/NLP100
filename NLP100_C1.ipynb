# 08

```python
def cipher(str):
    ciph = ''
    for c in str:
        if ord(c) in range(97, 122+1):
            ciph += chr(219 - ord(c))
        else:
            ciph += c
    return ciph

def decipher(str):
    return cipher(cipher(str))

import string

low_up = string.ascii_lowercase + string.ascii_uppercase
print(cipher(low_up))
print(decipher(low_up))
```

# 09

```python
import random

def change_order(str):
    sp = str.split()
    ans_L = []

    for w in sp:
        if len(w) > 4:
            mid = list(w[1:-1])
            random.shuffle(mid)
            res = w[0] + ''.join(mid) + w[-1]
            ans_L.append(res)
        else:
            ans_L.append(w)

    return ' '.join(ans_L)

s = 'I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind .'
change_order(s)
```
